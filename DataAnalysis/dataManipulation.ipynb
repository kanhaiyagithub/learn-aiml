```python
df['Unnamed: 10'] = pd.to_numeric(df['Unnamed: 10'], errors='coerce')
```

```python
## in one go 
df['Unnamed: 10_fillNA'] = pd.to_numeric(df['Unnamed: 10'], errors='coerce').fillna(
    pd.to_numeric(df['Unnamed: 10'], errors='coerce').mean()
)
```
