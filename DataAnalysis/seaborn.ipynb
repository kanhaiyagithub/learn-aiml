Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and in creating complex visulaization with just a few line of code.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
##Create a Scatter Plot 
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
##line plot 
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

```python
#Categorical plot 
#Seaborn allows for visualization of categorical variables such as sex, smoker, day, and time.
#Bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Plot of Total Bill by Day')
plt.show()
```

```python
##Box plot 
#Box plots are useful for identifying outliers and understanding the distribution of data. To create a box plot of total bill by day
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
##Violin plot 
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
#Histograms
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill with KDE')
plt.show()
```

```python
#KDE Plot 
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE Plot of Total Bill')
plt.show()
#The fill=True parameter fills the area under the KDE curve, providing a clearer visualization of the distribution.
```

```python
#Pair plot 
#Pair plots are useful for visualizing relationships between all numerical variables in the dataset
sns.pairplot(tips)
plt.show()
```

```python
#correlation and Heatmap 
#To visualize the correlation between numerical features, use a heatmap. First, select only the numerical columns and compute the correlation matrix
numerical_features = ['total_bill', 'tip', 'size']
correlation = tips[numerical_features].corr()
print(correlation)
#he correlation matrix shows the relationship between each pair of numerical features. For example, if total_bill increases by one unit, tip increases by approximately 0.67 units.
```

```python
#The heatmap uses color to represent the strength of the correlation between variables. Red indicates a strong positive correlation, while blue indicates a weaker or negative correlation.
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#We can customize plots further by combining Seaborn with matplotlib. For example, to plot total sales by product using a custom dataset:
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
#Plot total sales by product
plt.figure(figsize=(10, 6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.show()
```
